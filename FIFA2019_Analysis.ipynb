### Imputing missing values in selected columns for proper data visualisation

```python
fifa['Work Rate'].fillna('Medium/ Medium', inplace = True)
fifa['Work Rate'].value_counts(dropna = False)
```

```python
fifa['Skill Moves'].fillna(fifa['Skill Moves'].median(), inplace = True)
fifa['Skill Moves'].value_counts(dropna = True)
```

```python
fifa['Loaned From'].fillna('NA', inplace = True)
fifa['Loaned From'].value_counts(dropna = False)
```

```python
fifa['Body Type'].fillna('Normal', inplace = True)

# replacing the single body types with the Atlectic body types

fifa['Body Type'].replace('Messi', 'Atletic', inplace = True)
fifa['Body Type'].replace('C. Ronaldo', 'Atletic', inplace = True)
fifa['Body Type'].replace('Courtois', 'Atletic', inplace = True)
fifa['Body Type'].replace('PLAYER_BODY_TYPE_25', 'Atletic', inplace = True)
fifa['Body Type'].replace('Shaqiri', 'Atletic', inplace = True)
fifa['Body Type'].replace('Neymar', 'Atletic', inplace = True)
fifa['Body Type'].replace('Akinfenwa', 'Atletic',inplace = True)

fifa['Body Type'].value_counts(dropna = False)
```

```python
fifa['Club'].fillna('No Club', inplace = True)
fifa['Club'].value_counts(dropna = False)
```

##### Removing the € symbol and replaying K with 000 and M with 00000 to make it a Numeric Value

```python
fifa['Wage'].fillna('€200K', inplace = True)
fifa['Wage'].value_counts(dropna = False)
```

```python
# defining a function for cleaning the wage column

def extract_value_from(Value):
    out = Value.replace('€', '')
    if 'M' in out:
        out = float(out.replace('M', ''))*1000000
    elif 'K' in Value:
        out = float(out.replace('K', ''))*1000
    return float(out)
```

```python
# applying the function to the wage column
fifa['Value'] = fifa['Value'].apply(lambda x: extract_value_from(x))
fifa['Wage'] = fifa['Wage'].apply(lambda x: extract_value_from(x))
fifa['Wage']
```

```python
# Comparing the players' Wages

sns.set(style = 'dark', palette = 'bright', color_codes = True)
plt.figure(figsize = (75, 15))
x = fifa.Wage.head(100)
sns.countplot(x, data = fifa, color = 'darkblue')
plt.xlabel('Wage Range for Players', fontsize = 16)
plt.ylabel('Count of the Players', fontsize = 16)
plt.title('Comparing the wages of players', fontsize = 20)
plt.show()
```
